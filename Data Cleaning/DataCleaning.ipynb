```python
# Clean Age Variable
#Remove odd values

main_df["VicAge"] = main_df["VicAge"].replace(-3, 999)
main_df["VicAge"] = main_df["VicAge"].replace(-2,999)
main_df["VicAge"] = main_df["VicAge"].replace(-1,999)
main_df["VicAge"] = main_df["VicAge"].replace(0,999)
# Create AgeGroup Variable
bins = [0, 10, 18, 24, 39, 54, 69, 121, 999]
labels = ["Child", "Teen", "Young Adult", "25-39", "40-54", "55-69", "Senior 70+", "Unknown"]
main_df ["AgeGroup"]=pd.cut((main_df["VicAge"]), bins, labels=labels)
```

```python
# Recode VicRace into EEOC Categories
main_df["VicRace"] = main_df["VicRace"].replace({"-":" ", "A": "AANHPI", "B":"Black", "C":"AANHPI", "D": "AANHPI", "F":"AANHPI", "G":"AANHPI", "H":"Hispanic", "I":"AIAN", "J":"AANHPI", "K":"AANHPI", "L":"AANHPI", "O":"Multiracial", "P":"AANHPI", "S":"AANHPI", "U":"AANHPI", "V":"AANHPI", "W":"White", "X": " ", "Z":"AANHPI"}) 

```

```python
# Remove missing genders marked as X
main_df["VicSex"] = main_df["VicSex"].replace("X", None)
main_df["VicSex"] = main_df["VicSex"].replace("H", None)
main_df["VicSex"] = main_df["VicSex"].replace("-", None)
main_df["VicSex"] = main_df["VicSex"].replace("None", None)
main_df["VicSex"].unique()
```

```python
# Fix LocCode
import numpy as np
main_df["LocCode"] = main_df["LocCode"].replace(np.nan, "799")
main_df["LocCode"]=main_df["LocCode"].astype(str).astype(int)
```

```python
# Recode LocCode into a new variable called AddressType
address=[]
for value in main_df["LocCode"]:
    if value < 200:
        address.append("Public Space")
    elif value >200 and value <500:
        address.append("Buisness")
    elif value >500 and value<600:
        address.append("Residential")
    elif value>800:
        address.append("Public Transit")
    else:
        address.append("Other")
        
main_df["AddressType"]=address

# Drop LocCode
main_df=main_df.drop(columns = ["LocCode"])
```

```python
# Recode StatusCode into a new variable called ArrestMade
# Sort CRMCode into a CrimeType variable
SC =["AA", "AO", "CC", "IC", "JA", "JO"]
Arst=["Yes", "No", "No", "No", "Yes", "No"]

main_df["Arrest"]= main_df["StatusCode"].replace(SC, Arst)

```
